

### Q1. Create a vehicle class

To create a `Vehicle` class with an `__init__` method that initializes `name_of_vehicle`, `max_speed`, and `average_of_vehicle`, you can do the following:

```python
class Vehicle:
    def __init__(self, name_of_vehicle, max_speed, average_of_vehicle):
        self.name_of_vehicle = name_of_vehicle
        self.max_speed = max_speed
        self.average_of_vehicle = average_of_vehicle

    def display_info(self):
        return (f"Vehicle Name: {self.name_of_vehicle}, "
                f"Max Speed: {self.max_speed} km/h, "
                f"Average: {self.average_of_vehicle} km/l")
```

### Q2. Create a child class `Car` from the `Vehicle` class

The `Car` class inherits from `Vehicle` and adds a method to return the seating capacity:

```python
class Car(Vehicle):
    def seating_capacity(self, capacity):
        return f"The {self.name_of_vehicle} has a seating capacity of {capacity}."
```

### Q3. What is multiple inheritance? Write a Python code to demonstrate multiple inheritance.

**Multiple inheritance** occurs when a class inherits from more than one base class. This allows the child class to inherit attributes and methods from all parent classes.

Here's an example:

```python
class Engine:
    def engine_type(self):
        return "V8 Engine"

class Wheels:
    def wheel_type(self):
        return "All-terrain Wheels"

class Car(Engine, Wheels):
    def car_description(self):
        return (f"This car has a {self.engine_type()} and {self.wheel_type()}.")

# Example usage:
my_car = Car()
print(my_car.car_description())
```

In this example, the `Car` class inherits from both `Engine` and `Wheels` classes, and can use methods from both.

### Q4. What are getter and setter in Python? Create a class with getter and setter methods.

**Getters** and **setters** are methods used to access and modify the values of private attributes in a class.

Here's a class with getter and setter methods:

```python
class Person:
    def __init__(self, name):
        self.__name = name  # Private attribute

    # Getter method
    def get_name(self):
        return self.__name

    # Setter method
    def set_name(self, name):
        if isinstance(name, str) and name:
            self.__name = name
        else:
            print("Invalid name")

# Example usage:
person = Person("John")
print(person.get_name())  # Outputs: John
person.set_name("Doe")
print(person.get_name())  # Outputs: Doe
```

### Q5. What is method overriding in Python? Write a Python code to demonstrate method overriding.

**Method overriding** occurs when a subclass provides a specific implementation for a method that is already defined in its parent class. This allows a subclass to tailor or extend the behavior of methods inherited from the parent class.

Here's an example:

```python
class Animal:
    def speak(self):
        return "Animal speaks"

class Dog(Animal):
    def speak(self):
        return "Dog barks"

class Cat(Animal):
    def speak(self):
        return "Cat meows"

# Example usage:
animals = [Dog(), Cat()]

for animal in animals:
    print(animal.speak())  # Outputs: Dog barks \n Cat meows
```
