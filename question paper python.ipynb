### Q1. How do you comment code in Python? What are the different types of comments?

**Comments in Python** are used to explain and annotate your code. They are not executed by the interpreter. There are two main types of comments in Python:

1. **Single-Line Comments**: These are used to comment on a single line of code. They start with the `#` symbol.
   ```python
   # This is a single-line comment
   print("Hello, World!")  # This prints a greeting message
   ```

2. **Multi-Line Comments**: Python does not have a dedicated syntax for multi-line comments. Instead, multi-line comments are often written using a series of single-line comments or triple-quoted strings (`'''` or `"""`). Though triple-quoted strings are technically not comments, they are often used for multi-line comments or docstrings.
   ```python
   """
   This is a multi-line comment.
   It spans multiple lines.
   """
   print("Hello, World!")
   ```

### Q2. What are variables in Python? How do you declare and assign values to variables?

**Variables** in Python are names that refer to values stored in memory. They allow you to store, modify, and retrieve data. 

**Declaring and Assigning Values:**
In Python, you do not need to declare the type of the variable explicitly; it is determined automatically based on the assigned value.

**Example:**
```python
# Declare a variable and assign a value
age = 25            # integer variable
name = "Alice"      # string variable
height = 5.6        # float variable
is_student = True   # boolean variable
```

### Q3. How do you convert one data type to another in Python?

**Type conversion** in Python is done using built-in functions. Here are common conversions:

- **To Integer**: `int()`
  ```python
  num_str = "123"
  num_int = int(num_str)  # Converts string to integer
  ```

- **To Float**: `float()`
  ```python
  num_int = 10
  num_float = float(num_int)  # Converts integer to float
  ```

- **To String**: `str()`
  ```python
  num = 456
  num_str = str(num)  # Converts integer to string
  ```

- **To Boolean**: `bool()`
  ```python
  num = 0
  bool_value = bool(num)  # Converts number to boolean (0 is False, non-zero is True)
  ```

### Q4. How do you write and execute a Python script from the command line?

1. **Write a Python Script**: Create a file with a `.py` extension, such as `script.py`, and write your Python code in it.
   ```python
   # script.py
   print("Hello, World!")
   ```

2. **Execute the Python Script**:
   Open your command line (Terminal or Command Prompt) and run the script using the `python` command (or `python3` if using Python 3).
   ```sh
   python script.py
   ```

   If you are using Python 3, you might need to use:
   ```sh
   python3 script.py
   ```

### Q5. Given a list `my_list = [1, 2, 3, 4, 5]`, write the code to slice the list and obtain the sub-list `[2, 3]`.

**Slicing a List**: You can use list slicing to get a sub-list. Slicing is done using the `start:stop` syntax.

**Example:**
```python
my_list = [1, 2, 3, 4, 5]
sub_list = my_list[1:3]  # This gets elements from index 1 to 2 (3 is not included)
print(sub_list)  # Output: [2, 3]
```

### Q6. What is a complex number in mathematics, and how is it represented in Python?

**Complex Number**: In mathematics, a complex number has a real part and an imaginary part. It is generally expressed as \(a + bi\), where \(a\) is the real part and \(b\) is the imaginary part.

**Representation in Python**: Python has built-in support for complex numbers. You can represent complex numbers using the `j` suffix.

**Example:**
```python
# Complex number with real part 2 and imaginary part 3
c = 2 + 3j
print(c)  # Output: (2+3j)
```
