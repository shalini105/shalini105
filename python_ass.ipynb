Q3. Explain the use of the following operators using an example:
(i) /
(ii) %
(iii) //
(iv) **

Sure, let's explain each of these operators with examples:

### (i) `/` Operator (Division)

The `/` operator is used for division in Python. It divides the left operand by the right operand and returns a float value.

**Example:**
```python
# Division
result = 10 / 3
print(result)  # Output: 3.3333333333333335
```

In this example, `10 / 3` results in `3.3333333333333335`, which is a float because even though the operands are integers, division in Python always returns a float if the result isn't an exact integer.

### (ii) `%` Operator (Modulus)

The `%` operator returns the remainder of the division of the left operand by the right operand.

**Example:**
```python
# Modulus
remainder = 10 % 3
print(remainder)  # Output: 1
```

Here, `10 % 3` returns `1` because 10 divided by 3 gives a quotient of 3 with a remainder of 1.

### (iii) `//` Operator (Floor Division)

The `//` operator performs floor division, which divides the left operand by the right operand and rounds down to the nearest integer.

**Example:**
```python
# Floor division
result = 10 // 3
print(result)  # Output: 3
```

In this case, `10 // 3` returns `3`, which is the integer part of the division result `3.3333...`.

### (iv) `**` Operator (Exponentiation)

The `**` operator is used for exponentiation, where the left operand is raised to the power of the right operand.

**Example:**
```python
# Exponentiation
result = 2 ** 3
print(result)  # Output: 8
```

Here, `2 ** 3` calculates `2` raised to the power `3`, resulting in `8`.

### Summary

- `/`: Division operator, returns a float.
- `%`: Modulus operator, returns the remainder.
- `//`: Floor division operator, returns the integer part of the division.
- `**`: Exponentiation operator, raises the left operand to the power of the right operand.

This approach demonstrates how Python handles different data types within a list and how you can use a loop to inspect each element dynamically.

Q5. Using a while loop, verify if the number A is purely divisible by number B and if so then how many
times it can be divisible.

Certainly! To achieve this, we can use a `while` loop to repeatedly divide `A` by `B` until `A` is no longer divisible by `B`. Here's how you can implement this in Python:

```python
# Taking input for A and B
A = int(input("Enter the number A: "))
B = int(input("Enter the number B: "))

# Initializing a counter to count the number of divisions
count = 0

# Using a while loop to divide A by B as long as A is divisible by B
while A % B == 0:
    A = A // B  # Divide A by B and update A
    count += 1  # Increment the count

# Print the number of times A was divisible by B
print(f"A is divisible by B {count} times.")
```

### Example Explanation:
- Suppose `A = 100` and `B = 5`.
- Initially, `A` is divisible by `B` (`100 % 5 == 0`), so we enter the `while` loop.
- Inside the loop, `A` is divided by `B` (`100 // 5 = 20`), and `count` is incremented to `1`.
- The loop continues because `20 % 5 == 0` is true, so `A` becomes `20 // 5 = 4`, and `count` becomes `2`.
- The loop continues once more because `4 % 5 != 0` is false, so `A` remains `4`, and `count` remains `2`.
- Therefore, `A` was divisible by `B` 2 times.

### Note:
- The `while` loop continues as long as `A % B == 0`, ensuring that `A` is divisible by `B`.
- Each time `A` is divided by `B`, `A` is updated (`A = A // B`) and `count` is incremented to keep track of how many times `A` was divisible by `B`.
- If initially `A` is not divisible by `B` (`A % B != 0`), the loop will not execute at all, and `count` will remain `0`.

This method efficiently determines how many times `A` can be divided by `B` until it's no longer divisible, using a `while` loop.

Q7. What do you understand about mutable and immutable data types? Give examples for both showing
this property.

**Mutable and Immutable Data Types:**

In Python, objects are classified into two categories based on whether their values can change after they are created:
- **Mutable:** Objects whose values can be changed after creation are mutable.
- **Immutable:** Objects whose values cannot be changed after creation are immutable.

**Mutable Data Types:**
- Mutable objects can be modified after they are created.
- Examples of mutable data types in Python include lists, dictionaries, and sets.

**Example of Mutable Data Type:**

```python
# Example of a mutable list
my_list = [1, 2, 3, 4]
print("Original List:", my_list)

# Modifying the list
my_list[2] = 30
print("Modified List:", my_list)
```

Output:
```
Original List: [1, 2, 3, 4]
Modified List: [1, 2, 30, 4]
```

In this example, `my_list` is mutable because we can modify its elements. After creation, we change the third element from `3` to `30`.

**Immutable Data Types:**
- Immutable objects cannot be modified after they are created.
- Examples of immutable data types in Python include integers, floats, strings, tuples, and frozensets.

**Example of Immutable Data Type:**

```python
# Example of an immutable tuple
my_tuple = (10, 20, 30)
print("Original Tuple:", my_tuple)

# Trying to modify the tuple (will result in an error)
# my_tuple[1] = 25  # Uncommenting this line will raise an error

# Even though we can create a new tuple based on the original tuple
my_tuple = (10, 25, 30)
print("Modified Tuple:", my_tuple)
```

Output:
```
Original Tuple: (10, 20, 30)
Modified Tuple: (10, 25, 30)
```

In this example, `my_tuple` is immutable. Trying to modify an element of the tuple directly (`my_tuple[1] = 25`) will result in a `TypeError` because tuples do not support item assignment after creation. Instead, we can create a new tuple based on the original tuple.

### Summary:
- **Mutable** objects can be changed after creation (e.g., lists, dictionaries, sets).
- **Immutable** objects cannot be changed after creation (e.g., integers, floats, strings, tuples).
- Understanding mutability versus immutability is important in Python because it affects how you manipulate and work with data, especially when dealing with function arguments, data sharing, and memory management.
